```python
def StuckTuples(data, n): 
    return n - data["id"].count()
```

```python
def GetStuckIds(data, n):
    allIds = pd.DataFrame(np.arange(n))
    ids = data["id"].as_matrix()
    #print(allIds.isin(ids).all(axis=1))
    stuckIds = allIds[~allIds.isin(ids).all(axis=1)]
    return stuckIds

    
def StuckDistribution(data, n, bins = 50):
    stuckIds = GetStuckIds(data, n)
    # stuckIds.plot.hist(bins = 100)
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.hist(stuckIds.as_matrix(), bins = bins)
    ax.set(title='Stuck Tuples, Frequency by Ingestion Order, {} bins.'.format(bins),
            xlabel = "Tuple Ingestion Order",
            ylabel="Frequency")
    plt.show()

StuckDistribution(data, n)
```

```python
## For different N (constant k,l,p) get the number of stuck tuples. 
k, l, p = 40, 5, 10
dataframes, stuck, stuckPercentage = [], [], []
n_values = [50000, 250000, 500000, 750000, 1000000]

for i,n in enumerate(n_values): 
    data = ReadFile(k, l, p, n, "ntest")
    s = StuckTuples(data, n)
    #print("Length: {}, N: {}".format(data.count()[0], n))
    stuck.append(s)
    stuckPercentage.append(round(s/n*100, 2))

    print("{} out of {} tuples stuck. ({}%) ".format(s, n, round(s/n*100, 2)))
    

print(stuck)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(n_values, stuckPercentage, linewidth=3)
ax.scatter(n_values, stuckPercentage, marker = "o")
ax.set(title='Stuck tuples vs. Datastream Length',
        xlabel = "Datastream Length",
        ylabel="Stuck Tuples (%)",
      ylim = (0, 5))
plt.show()


fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(n_values, stuck, linewidth=3)
ax.scatter(n_values, stuck, marker = "o")
ax.set(title='Stuck tuples vs. Datastream Length',
        xlabel = "Datastream Length",
        ylabel="Stuck Tuples",
      ylim = (0, 5000))
plt.show()

StuckDistribution(data, 1000000, 50)
```
